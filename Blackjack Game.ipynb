```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
    def __str__(self):
        return (f'{self.rank} of {self.suit}')
```

```python
import random
class Deck():
    def __init__(self):
        self.allcards=[]
        for suit in suits:
            for rank in ranks:
                self.allcards.append(Card(suit,rank))
    def Shuffle(self):
        random.shuffle(self.allcards)
    def deal_one(self):
        return self.allcards.pop()
```

```python
class Player:
    def __init__(self):
        y=input("Enter name of player: ")
        self.name=y
        a=int(input("Enter your total chips: "))
        self.total=a
        while True:
            b=int(input("Enter your betting chips: "))
            if b<a:
                self.bet=b
                break
            else:
                print("enter correct betting amount")
        self.playercards=[]
        self.sum=0
    def add_cards(self,card):
        self.playercards.append(card)
        self.sum+=card.value
            #Ace value 1 or 11 condition
        self.aces=0
        if card.rank=='Ace':
            self.aces+=1
        while self.sum>21 and self.aces:
            self.sum-=10
            self.aces-=1
    def printcards(self):
        print(f'\n\n{self.name} cards are:')
        for i in self.playercards:
            print(i) 
        print(f"Total value of cards are: {self.sum}")
    def __str__ (self):
        return f'player {self.name} total amount of {self.total}'
```
