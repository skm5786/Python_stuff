```python
import random
import itertools
from IPython.display import clear_output
```

```python
def TicTacToe():
  board = [1, 2, 3, 4, 5, 6, 7, 8, 9]
  end = False
  MagicSquare = [8,3,4,1,5,9,6,7,2]

  def PrintBoard():
    print()
    print('', board[0], "|", board[1], "|", board[2])
    print("---|---|---")
    print('', board[3], "|", board[4], "|", board[5])
    print("---|---|---")
    print('', board[6], "|", board[7], "|", board[8])
    print()

  def GetNumber():
    while True:
      number = input()
      try:
        number  = int(number)
        if number in range(1, 10):
          return number
        else:
          print("\nNumber not on board")
      except ValueError:
        print("\nThat's not a number. Try again")
        continue

  def Turn(player):
    placing_index = GetNumber() - 1
    if board[placing_index] == "X" or board[placing_index] == "O":
      print("\nBox already occupied. Try another one")
      Turn(player)
    else:
      board[placing_index] = player

  def CheckWin(player):
    count = 0

    for x in range(9):
      for y in range(9):
        for z in range(9):
          if x != y and y != z and z != x:
            if board[x] == player and board[y] == player and board[z] == player:
              if MagicSquare[x] + MagicSquare[y] + MagicSquare[z] == 15:
                print("Player", player ,"wins!\n")
                return True

    for a in range(9):
      if board[a] == "X" or board[a] == "O":
        count += 1
      if count == 9:
        print("The game ends in a Tie\n")
        return True
  def computer_turn(count):
    if count==1:
        kk=random.randrange(0,8)
        board[kk]="X"
        PrintBoard()
        return False
    if count==2: 
        while True:
            try:    
                kk=random.randrange(0,9)
                if board[kk]!='X' and board[kk]!='O':
                    board[kk]="X"
                    PrintBoard()
                    return False
            except:
                continue

    if count>2:
    #checking for computer win
        o_places=[]
        for x,n in enumerate(board):
            if n=="O":
                o_places.append(x)
        x_places=[]
        for x,n in enumerate(board):
            if n=="X":
                x_places.append(x)
        x_combinations=list(itertools.combinations(x_places,2))
        o_combinations=list(itertools.combinations(o_places,2))
        k=0
        km=0
        for x,y in x_combinations:
            m=15-(MagicSquare[x]+ MagicSquare[y])
            for a,n in enumerate(MagicSquare):
                if n==m:
                    k=a
            if m<9 and m>=0 and (board[k] != "O" and board[k] != "X"):
                    board[k]="X"
                    print('computer wins')
                    PrintBoard()
                    return  True
            elif m<9 and m>=0 and (board[k] == "O" or board[k] == "X"):
                    #chance of winning of human
                    for x,y in o_combinations:
                        m=15-(MagicSquare[x]+ MagicSquare[y])
                        for a,n in enumerate(MagicSquare):
                            if n==m:
                                km=a
                        if board[km] != "X" and board[km] != "O": 
                            board[km]="X"
                            PrintBoard()
                            return False 
        while True:
                try:    
                    kk=random.randrange(0,8)
                    if kk not in x_places and kk not in o_places:
                        board[kk]="X"
                        PrintBoard()
                        break
                except:
                        continue        
        for a in range(9):
            if board[a] == "X" or board[a] == "O":
                count += 1
            if count == 9:
                print("The game ends in a Tie\n")
                return True
        return False        
                
  count=1              
  while True:
    clear_output()    
    PrintBoard()
    end = CheckWin("O")
    if end == True:
      break
    print("computer turn")

    end=computer_turn(count)
    count+=1
    if end == True:
      break
    print("Choose a box player O")
    Turn("O")
    


    
        
```
